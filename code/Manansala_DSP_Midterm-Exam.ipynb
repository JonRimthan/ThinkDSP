g.Print the highest points in the spectrum and their frequencies, in descending order.

```python
spectrum1 = segment1.make_spectrum()
spectrum2 = segment2.make_spectrum()
spectrum3 = segment3.make_spectrum()
```

```python
spectrum1.plot(high=3000)
plt.xlabel('Frequency (Hz)');
```

```python
spectrum2.plot(high=3000)
plt.xlabel('Frequency (Hz)');
```

```python
spectrum3.plot(high=3000)
plt.xlabel('Frequency (Hz)');
```

```python
#Audio 1 Highest Points in descending order

# Find the highest amplitudes and their corresponding frequencies
amplitudes1 = spectrum1.amps.ravel()
frequencies1 = spectrum1.fs.ravel()
indices1 = np.argsort(amplitudes1)[::-1]

# Print the highest amplitudes and their corresponding frequencies
for i in indices1[:10]:
    print(f"Amplitude: {amplitudes1[i]}, Frequency: {frequencies1[i]}")
```

```python
#Audio 2 Highest Points in descending order

# Find the highest amplitudes and their corresponding frequencies
amplitudes2 = spectrum2.amps.ravel()
frequencies2 = spectrum2.fs.ravel()
indices2 = np.argsort(amplitudes2)[::-1]

# Print the highest amplitudes and their corresponding frequencies
for i in indices2[:10]:
    print(f"Amplitude: {amplitudes2[i]}, Frequency: {frequencies2[i]}")
```

```python
#Audio 3 Highest Points in descending order

# Find the highest amplitudes and their corresponding frequencies
amplitudes3 = spectrum3.amps.ravel()
frequencies3 = spectrum3.fs.ravel()
indices3 = np.argsort(amplitudes3)[::-1]

# Print the highest amplitudes and their corresponding frequencies
for i in indices3[:10]:
    print(f"Amplitude: {amplitudes3[i]}, Frequency: {frequencies3[i]}")
```
